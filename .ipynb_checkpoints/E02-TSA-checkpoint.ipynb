# 02.1 Find the values of rho for which a series becomes non-stationary. rho between 0 and 1 ([0,1]) \n",
`X(t) = Rho * X(t-1) + Er(t)`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller

%matplotlib inline
sns.set()
```

a) Definimos la función con la que se crea el dataset

```python

def ts(rho):
    ts = pd.Series(np.random.rand(2000) - 0.5, index=pd.date_range("2020-01-01", periods=2000))
    
    for i in range(1, ts.shape[0]):
        ts.iloc[i] = ts.iloc[i] + rho * ts.iloc[i-1]    
    return ts
    
```

b) Se construye el loop para encontrar el rho donde inicia a rechazarse Ho, ya que se desea encontrar a partir de que valor la seríe no sería estacionaria

```python

for i in range(101):
    print(i/100)
    print('p-value: %f' % adfuller(ts(rho=i/100))[1])
```

Respuesta: A medida que Rho sea mas cercano a 1, el test de dicker fuller no rechaza su hipótesis nula, por lo tanto, la serie NO ES ESTACIONARIA.

# 02.2 Download and explore two series from google trends

Se tomaron desde google trends la series correspondientes a las busquedas de las palabras "iphone" y "android" a nivel mundial desde el año 2015, como se muestr a continuación:

```python
iphone = pd.read_csv('datasets\iphone.csv', sep=',', index_col='Semana')
```

```python
iphone.info()
```

```python
iphone.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

a) Luego de graficar las series iniciamos aplicando la prueba de Dickey Fuller con el ánimo de identificar si estas son estacionarias

```python
result = adfuller(iphone['iPhone'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

- Para el caso de la serie de "iphone" se encuentra que se rechaza la Ho por lo cual se tiene evidencia estadistica suficiente para afirmar que la seríe es estacionaria sin aplicarle ningun tipo de transformación

```python
result = adfuller(iphone['android'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

- Por el contrario, en el caso de la serie "android" encontramos no se rechaza Ho, por lo cual no se tiene evidencia estadistica suficiente para decir que esta sea estacionaria.

b) Teniendo en cuenta las conclusiones del punto anterior se encuentra necesario identificar las tendencias en las series:

```python
iphone.rolling(52).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

- Luego de aplicar una media movil de 52 semanas, se identifica que en la serie de "Android" hay una tendencia a la baja, mientras que por su parte, la serie de "iphone" tiene una tendencia al alza

c) Teniendo en cuenta lo anterior, se aplica una diferenciacion de las series para eliminar el efecto de sus tendencias

```python
iphone.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
print('p-value test Dickey Fuller para iphone: %f' % adfuller(iphone['iPhone'])[1])
print('p-value test Dickey Fuller para android: %f' % adfuller(iphone['android'])[1])
```

```python
print('p-value test Dickey Fuller para iphone diferenciada: %f' % adfuller(iphone['iPhone'].diff().iloc[1:])[1])
print('p-value test Dickey Fuller para android diferenciada: %f' % adfuller(iphone['android'].diff().iloc[1:])[1])
```

- Luego de tener la series diferenciadas, encontramos que para ambos casos se rechaza Ho, es decir, se tiene evidencia estadistica suficiente para mencionar que son estacionarias

d) Con el ánimo de verificar si el comportamiento de las series es similar, se construyen las matrices de correlación tanto para la series sin transformaciones y diferenciadas

```python
iphone.corr()
```

```python
iphone.diff().corr()
```

- Teniendo en cuenta los resultados anteriores se encuentra que la máxima correlación se tiene cuando las series se encuentran diferenciadas, no obstante se tiene que esta es del 23.92%, lo cual se entiende que se debe a el comportamiento que tienen antes del 2016-06-05, y a partir de allí su comportamiento es distinto

e) Finalmente, se construyen los graficos de autocorrelación de las series

```python
plt.figure(figsize=(20,10))
pd.plotting.autocorrelation_plot(iphone['iPhone']);
```

- Para el caso de la serie de "iphone", tenemos que sin diferenciar la serie existe una autorrelación de la misma cada 52 semanas aproximadamente

```python
plt.figure(figsize=(20,10))
pd.plotting.autocorrelation_plot(iphone['android']);
```

- Para el caso de "android", aunque no es estancionaria se encuentra que la autocorrelación decrece con el tiempo
